```
import numpy as np
import pylab
import scipy.stats as stats
```

```
std_normal=np.random.normal(loc=0, scale=1, size=1000)

for i in range(0,101):
  print(i, np.percentile(std_normal, i))
```

```
std_normal=np.random.normal(loc=1, scale=2, size=100)

for i in range(0,10):
  print(i , np.percentile(std_normal, i))
```

```
meas=np.random.normal(loc=20, scale=5, size=100)

stats.probplot(meas, dist="norm", plot=pylab, fit=True, rvalue=False)
pylab.show()
```

```
meas=np.random.normal(loc=30, scale=2.5, size=1000)

stats.probplot(meas, dist="norm", plot=pylab)
pylab.show()
```
